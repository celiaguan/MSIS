```python
import pandas as pd
import time
import seaborn as sns
import matplotlib.pyplot as plt
import datetime
import numpy as np

from functools import reduce
```

```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
```

```python
vol = pd.read_csv('bitcoin volume.csv').rename(columns=lambda x: x.strip())
vol=vol.fillna(0)
```

```python
# convert the date string to the correct date format
vol.Time = pd.to_datetime(vol['Time'])
```

```python
vol.head()
```

```python
# pivot the table 
vol =pd.melt(vol, id_vars=['Time'], value_vars=['mtgox','bitfinex','bitflyer','bitstamp','btcchina','btce','huobi','lakebtc','okcoin','others'])
vol =vol.rename(columns={"variable":"exchange","value":"vol_total"})
vol.head()
```

```python
cny = pd.read_csv('CNY.csv').rename(columns=lambda x: x.strip())
cny.Time = pd.to_datetime(cny['Time'])
cny=cny.fillna(0)
cny.head()
```

```python
# pivot the table 
cny =pd.melt(cny, id_vars=['Time'], value_vars=['mtgox','btcchina','huobi','lakebtc','localbitcoins','okcoin','rmbtb'])
cny =cny.rename(columns={"variable":"exchange","value":"vol_cny"})
cny.head()
```

```python
eur = pd.read_csv('EUR.csv').rename(columns=lambda x: x.strip())
eur.Time = pd.to_datetime(eur['Time'])
eur=eur.fillna(0)
eur.head()
```

```python
# pivot the table 
eur =pd.melt(eur, id_vars=['Time'], value_vars=['mtgox','bix-x','itbit','bitcoin24','bitstamp','bitcoinde','coinbase','gatecoin','kraken','others'])
eur =eur.rename(columns={"variable":"exchange","value":"vol_eur"})
eur.head()
```

```python
inr = pd.read_csv('INR.csv').rename(columns=lambda x: x.strip())
inr.Time = pd.to_datetime(inr['Time'])
inr.head()
```

```python
# pivot the table 
inr=pd.melt(inr, id_vars=['Time'], value_vars=['localbitcoins'])
inr =inr.rename(columns={"variable":"exchange","value":"vol_inr"})
inr.head()
```

```python
jpy = pd.read_csv('JPY.csv').rename(columns=lambda x: x.strip())
jpy.Time = pd.to_datetime(jpy['Time'])
jpy=jpy.fillna(0)
jpy.head()
```

```python
# pivot the table 
jpy =pd.melt(jpy, id_vars=['Time'], value_vars=['mtgox','anxbtc','bitflyer','kraken','localbitcoins'])
jpy =jpy.rename(columns={"variable":"exchange","value":"vol_jpy"})
jpy.head()
```

```python
# pivot the table 
krw =pd.melt(krw, id_vars=['Time'], value_vars=['kraken'])
krw =krw.rename(columns={"variable":"exchange","value":"vol_krw"})
krw.head()
```

```python
sgd = pd.read_csv('SGD.csv').rename(columns=lambda x: x.strip())
sgd.Time = pd.to_datetime(sgd['Time'])
sgd=sgd.fillna(0)
sgd
```

```python
# pivot the table 
sgd =pd.melt(sgd, id_vars=['Time'], value_vars=['mtgox','anxbtc','itbit','localbitcoins'])
sgd =sgd.rename(columns={"variable":"exchange","value":"vol_sgd"})
sgd.head()
```

```python
usd = pd.read_csv('USD.csv').rename(columns=lambda x: x.strip())
usd.Time = pd.to_datetime(usd['Time'])
usd=usd.fillna(0)
usd.head()
```

```python
# pivot the table 
usd =pd.melt(usd, id_vars=['Time'], value_vars=['mtgox','bix-x','itbit','bitfinex','bitstamp','btce','coinbase','lakebtc','okcoin','others'])
usd =usd.rename(columns={"variable":"exchange","value":"vol_usd"})
usd.head()
```

```python
vols=[vol,cny,usd,eur,inr,jpy,sgd]
vol_final = reduce(lambda left,right: pd.merge(left,right,on=['Time','exchange'],how='outer'), vols)
```

```python
vol1 = pd.merge(vol,cny,on=['Time','exchange'],how='outer')
```

```python
vol_final=vol_final.fillna(0).sort_values(by='Time')
vol_final=vol_final[vol_final['Time'].dt.year>2009]
vol_final
```

```python
vol_final.to_csv('vol_clean.csv')
```
